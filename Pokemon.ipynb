## Challenge 3: Grouping and Aggregation
1. Group the PokÃ©mon by their primary type and find the average attack for each type.
2. Find the maximum defense stat for each PokÃ©mon type.


```python
#Group the PokÃ©mon by their primary type and find the average attack for each type.
average_attack_by_type = df.groupby('Type 1')['Attack'].mean()
average_attack_by_type
```

```python
# the maximum defense stat for each PokÃ©mon type.
max_defense = df.groupby('Type 1')['Defense'].max()
max_defense
```

```python
# Combine average attack and max defense into one DataFrame
stats_by_type = df.groupby('Type 1').agg({'Attack': 'mean', 'Defense': 'max'})

# Rename columns for clarity
stats_by_type = stats_by_type.rename(columns={'Attack': 'Avg Attack', 'Defense': 'Max Defense'})

# Display the combined DataFrame
stats_by_type
```

## Challenge 5: Handling Missing Data
1. Identify if there are any missing values in the dataset.
2. Replace any missing values with the column mean or median.

```python
#Identifying Missing values
print("Missing Data Summary:")
missing_data = df.isnull().sum()
missing_data
```

```python
# Fill missing values in numeric columns with their mean
df_filled_mean = df.copy()  # Make a copy to avoid modifying the original DataFrame
df_filled_mean = df_filled_mean.apply(
    lambda col: col.fillna(col.mean()) if col.dtype in ['float64', 'int64'] else col
)

# Display the updated DataFrame
df_filled_mean.head()
```

```python
# Fill missing values in numeric columns with their median
df_filled_median = df.copy()  # Make a copy of the DataFrame to avoid modifying the original
df_filled_median = df_filled_median.apply(
    lambda col: col.fillna(col.median()) if col.dtype in ['float64', 'int64'] else col
)

# Display the updated DataFrame
df_filled_median.head()
```

```python
# Fill missing categorical values with a placeholder
df['Type 2'] = df['Type 2'].fillna('Unknown')

# Display the updated DataFrame
df
```

```python
# Verify if there are any missing values left in the dataset
missing_values_after = df.isnull().sum()

# Display the result
missing_values_after
```

# Challenge 6: Adding New Columns
1. Create a new column called `Power Ratio` as the ratio of attack to defense.
2. Add a column called `Is_Legendary` that is `True` if the PokÃ©mon is legendary, otherwise `False`.

```python
# Add a new column 'Power Ratio' as the ratio of Attack to Defense
df['Power Ratio'] = df['Attack'] / df['Defense']

# Display the DataFrame with the new column
df.head()
```

```python
# Ensure the 'Is_Legendary' column exists or is set based on the 'Legendary' column
df['Is_Legendary'] = df['Legendary'] == True

# Display the DataFrame with the new column
df.head()
```

```python
# Add a new column 'Total Stats' as the sum of all stat columns
df['Total Stats'] = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].sum(axis=1)

# Display the DataFrame with the new column
df.head()
```

## Challenge 7: Advanced Analysis
1. Find the PokÃ©mon with the highest combined attack and speed.
2. For each type, find the PokÃ©mon with the best defense.

```python
# Add a column for combined Attack and Speed
df['Attack + Speed'] = df['Attack'] + df['Speed']

# Find the PokÃ©mon with the highest combined Attack and Speed
highest_attack_speed = df.loc[df['Attack + Speed'].idxmax()]

# Display the PokÃ©mon with the highest combined Attack and Speed
highest_attack_speed
```

```python
# Group by 'Type 1' and find the PokÃ©mon with the maximum Defense for each type
best_defense_per_type = df.loc[df.groupby('Type 1')['Defense'].idxmax()]

# Display the DataFrame with the best Defense PokÃ©mon for each type
best_defense_per_type[['Type 1', 'Name', 'Defense']]
```

# Challenge 8: Visualization
1. Plot a bar chart showing the average HP for each PokÃ©mon type.
2. Create a scatter plot of Attack vs. Defense for all PokÃ©mon.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate average HP for each Type 1
avg_hp_by_type = df.groupby('Type 1')['HP'].mean().sort_values()

# Plot the bar chart
plt.figure(figsize=(10, 6))
avg_hp_by_type.plot(kind='bar', color='skyblue')
plt.title('Average HP by PokÃ©mon Type', fontsize=14)
plt.xlabel('Type 1', fontsize=12)
plt.ylabel('Average HP', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Scatter plot of Attack vs. Defense
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Attack', y='Defense', hue='Type 1', palette='Set2', alpha=0.8)
plt.title('Attack vs. Defense for PokÃ©mon', fontsize=14)
plt.xlabel('Attack', fontsize=12)
plt.ylabel('Defense', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Type 1')
plt.tight_layout()
plt.show()
```
